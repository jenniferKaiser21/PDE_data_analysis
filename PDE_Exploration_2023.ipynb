```python
"""
Finding counts of records in "YearsInEd" column of df_13_14
"""
yearsined_1314 = df_13_14['YearsInEd'].value_counts().sort_index(ascending=True)
df_13_14['YearsInEd'].value_counts().sort_index(ascending=True)
```

```python
"""
Finding counts of records in "YearsInED" column of df_21_22
"""
yearsined_2122 = df_21_22['YearsInED'].value_counts().sort_index(ascending=True)
df_21_22['YearsInED'].value_counts().sort_index(ascending=True)
```

```python
"""
Comparing yearsined_1314 to yearsined_2122
"""
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.bar(yearsined_1314.index, yearsined_1314.values)
ax1.set_title("Number of Years in Education 2013-2014")
ax1.set_xlabel("Number of Years in Education")
ax1.set_ylabel("Number of Educators")

ax2.bar(yearsined_2122.index, yearsined_2122.values)
ax2.set_title("Number of Years in Education 2021-2022")
ax2.set_xlabel("Years in Education")
ax2.set_ylabel("Number of Years in Education")

plt.show()
```

```python
"""
Comparing yearsined_1314 to yearsined_2122 in the same bar chart
"""
df_counts = pd.concat([yearsined_1314, yearsined_2122], axis=1, keys=['13-14', '21-22'])
colors = ['blue', 'red']
df_counts.plot(kind='bar', figsize=(20,10), color=colors)
plt.title("Comparing Years in Education, 2013-2014 vs 2021-2022")
plt.xlabel("Number of Years in Education")
plt.ylabel("Number of Educators")
plt.show()
```
